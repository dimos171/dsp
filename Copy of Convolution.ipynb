```python
import mysignals as sigs
from matplotlib import pyplot as plt
from matplotlib import style
from scipy import signal
import csv

#1
output_signal = signal.convolve(sigs.InputSignal_1kHz_15kHz, sigs.Impulse_response, mode='same')

#2
def convolution(sig_src_arr,imp_response_arr,sig_dest_arr):
  for x in range(len(sig_src_arr)+len(imp_response_arr)):
    sig_dest_arr[x] = 0
  #plt.rcParams["figure.figsize"] = (20,3)
  for x in range(len(sig_src_arr)):
    #print("X =",x)
    for y in range(len(imp_response_arr)):
      val = sig_dest_arr[x+y] + sig_src_arr[x]*imp_response_arr[y]
      #print("[" + str(x+y) + "] = " + str(sig_dest_arr[x+y]) + " + "  + str(sig_src_arr[x]) + " * "  + str(imp_response_arr[y]) + " = " + str(val))
      sig_dest_arr[x+y] = val
    #print(sig_dest_arr)  
    #plt.plot(sig_src_arr, color='pink')
    #plt.plot(sig_dest_arr)
    #plt.show()
    #print("")
  

#input = [0,1,2,1,0,-1,-2,-1,0,1,2,1,0,-1,-2,-1,0,1,2,1,0,-1,-2,-1,0] 
#input = [0,1,2,3,4,5,6,7,8]
#input = [1,1,1,1,1,1,1,1,1]
#imp = [1,-1]
input=sigs.InputSignal_1kHz_15kHz
imp=sigs.Impulse_response
output_signal = [None]*(len(input)+len(imp))

convolution(input, imp, output_signal)

#csvfile = "convolution_output.txt"
#with open(csvfile,"w") as output:
#  writer = csv.writer(output,lineterminator=',')
#  for x in output_signal:
#    writer.writerow([x])

style.use('ggplot')

f,plt_arr = plt.subplots(3,sharex=True)

plt_arr[0].plot(input)
plt_arr[0].set_title("Input Signal")
plt_arr[1].plot(imp, color='brown')
plt_arr[1].set_title("Impulse")
plt_arr[2].plot(output_signal, color='green')
plt_arr[2].set_title("Output Signal")

plt.rcParams["figure.figsize"] = (20,10)
plt.show()





```
