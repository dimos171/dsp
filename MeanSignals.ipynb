```python
import numpy as np
import mysignals as sigs

#1
signal_mean = np.mean(sigs.InputSignal_1kHz_15kHz)
print("Numpy Mean: ",signal_mean)

variance = np.var(sigs.InputSignal_1kHz_15kHz)
print("Numpy Variance: ",variance)

standard_deviation = np.std(sigs.InputSignal_1kHz_15kHz)
print("Numpy Standard Deviation: ",standard_deviation)

#2
_mean = 0.0
_variance = 0.0
_std = 0.0

def calc_mean(sig_src_arr):
  global _mean
  for x in range (len(sig_src_arr)):
    _mean += sig_src_arr[x]
  _mean = _mean/len(sig_src_arr)
  return _mean

print("Mean", calc_mean(sigs.InputSignal_1kHz_15kHz))

def calc_variance(sig_src_arr):
  global _mean
  global _variance
  _mean = calc_mean(sig_src_arr)

  for x in range (len(sig_src_arr)):
    _variance = _variance + (sig_src_arr[x] - _mean)**2
  _variance = _variance/len(sig_src_arr)
  return _variance

print("Variance", calc_variance(sigs.InputSignal_1kHz_15kHz))

def calc_standard_deviation(sig_src_arr):
  global _variance
  global _std
  _variance = calc_variance(sig_src_arr)
  _std = _variance**(.5)
  return _std

print("Standard Deviation", calc_standard_deviation(sigs.InputSignal_1kHz_15kHz))



```
